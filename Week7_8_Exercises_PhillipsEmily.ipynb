# Checking Missing Data

# Filling Missing Data

```python
#using the fillna ffill method to fill missing data in with previous valid observation in the same column
data_candy_df = data_candy_df.fillna(method='ffill')
```

```python
#dropping first row in the dataframe as it contains mostly missing values
data_candy_df = data_candy_df.iloc[1:,:]
```

```python
#checking for columns with any missing data again
print("Columns with missing data:\n" + str(data_candy_df.columns[data_candy_df.isnull().any()]))
print("5 columns are still missing data")
```

# Filter Out Missing Data

```python
#dropping the 5 columns which still have missing data 
data_candy_df = data_candy_df.dropna(axis=1)
```

```python
data_candy_df.shape
print("2459 rows and 115 columns")
print("5 columns have been dropped for having missing data")
```

# Looking at Indices

# Hierarchical Indexing

```python
#setting the descriptive fields in the dataframe as the indices
#ID, Gender, Age, Country and State
multi = data_candy_df.set_index(['Internal ID','Q2: GENDER','Q3: AGE','Q4: COUNTRY','Q5: STATE, PROVINCE, COUNTY, ETC'])
```

```python
multi.head()
```

The dataframe is now indexed by the variables which define each questionaire which are: ID, gender, age, country and state/provice/county. These will help us identify each record, and give more understanding to each row in terms of survey responses. 

```python
multi.index.names
```

# Reshape

```python
#now that indices have been set, we can use the stack() method to pivot the columns into rows
result = multi.stack()
```

```python
result
```

By pivoting the dataframe's columns into rows, it produces a Series where each group pertains to the survey responses for each respondent based on their ID, gender, age, country and state/province/county.

It helps organize all of the data, and gives a better idea of the responses individually. 

# Other Dataset

```python
#MetObjects csv file -- > read into df
met_objects = pd.read_csv("Week7_8/MetObjects.csv")
met_objects.head()
```

```python
met_objects.shape
```

```python
met_objects.dtypes
```

# Dropping Missing & Duplicates

```python
#drop any rows with all missing data
met_objects = met_objects.dropna(how='all')
```

```python
print(met_objects.shape)
print("21 rows dropped for having all NA values")
```

```python
#drop any columns with all missing data
met_objects = met_objects.dropna(axis=1,how='all')
```

```python
print(met_objects.shape)
print("3 columns dropped")
```

```python
#dropping duplicates
met_objects = met_objects.drop_duplicates()
```

```python
print(met_objects.shape)
print("165 duplicate rows dropped")
```

# Split/Apply/Combine

#### Split

```python
#split the data into groups
#each group is the set of objects in each department
df_by_department = met_objects.groupby('Department')
df_by_department.head()
```

#### Apply

```python
#applying the describe function
#summary stats over departments
df_by_department.describe().head()
```

#### Combine

```python
#getting the count for number of values in each column for each department group
#combining into a new dataframe
df_count_by_department = df_by_department.count()
df_count_by_department.head()
```

# Cross Tabs

```python
#summarize the data by object classification and culture
pd.crosstab([met_objects['Classification']],met_objects['Culture'],margins=True)
```

# Converting Between String and Datetime

```python
#convert dates in table to actual date time format --> they were previously object types
#errors = 'coerce' will mark any invalid dates as NaT
#convert not parseable values to string
met_objects['Artist Begin Date'] =  pd.to_datetime(met_objects['Artist Begin Date'].astype(str),errors='coerce')
```

```python
met_objects['Artist End Date'] =  pd.to_datetime(met_objects['Artist End Date'].astype(str),errors='coerce')
```

```python
met_objects['Object Date'] =  pd.to_datetime(met_objects['Object Date'],errors='coerce')
```

```python
met_objects['Object Begin Date'] =  pd.to_datetime(met_objects['Object Begin Date'],errors='coerce')
```

```python
met_objects['Object End Date'] =  pd.to_datetime(met_objects['Object End Date'],errors='coerce')
```

```python
met_objects['Metadata Date'] =  pd.to_datetime(met_objects['Metadata Date'],errors='coerce')
```

```python
met_objects.dtypes
```

# Date Offsets

```python
#finding the Monday or next Monday for each start date --> start of the week
#if the Object Begin Date is already a Monday, the new column will have the same value
#if not, it will find the date for the next Monday
met_objects['mondayDate'] = met_objects['Object Begin Date'].apply(lambda x: x + pd.DateOffset(days=7-x.weekday()) if  x.weekday() else x)
```
